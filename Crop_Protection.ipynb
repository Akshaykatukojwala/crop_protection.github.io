```python
from keras.layers import Conv2D
from keras.layers import MaxPooling2D
from keras.layers import Flatten
from keras.models import Sequential
from keras.layers import Dense
```

```python
crop_protection = Sequential()
```

```python
crop_protection.add(Conv2D(32, 3, 3, input_shape = (64, 64, 3), activation = "relu"))
```

```python
crop_protection.add(MaxPooling2D(pool_size = (2,2)))
```

```python
crop_protection.add(Flatten())
```

```python
crop_protection.add(Dense(output_dim = 128, activation = "relu", init = "random_uniform"))
```

```python
crop_protection.add(Dense(output_dim = 3, activation = "softmax", init = "random_uniform"))
```

```python
crop_protection.compile(optimizer = "adam", loss = "categorical_crossentropy", metrics = ["accuracy"]) 
```

```python
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator(rescale = 1./255, shear_range = 0.2, zoom_range = 0.2, horizontal_flip = True)
test_datagen = ImageDataGenerator(rescale = 1./255)
```

```python
x_train = train_datagen.flow_from_directory(r"C:\Users\Saicharan Pogul\Desktop\SmartBridge\Project\dataset\training_set", target_size = (64, 64), batch_size = 32, class_mode = "categorical")
```

```python
x_test = train_datagen.flow_from_directory(r"C:\Users\Saicharan Pogul\Desktop\SmartBridge\Project\dataset\test_set", target_size = (64, 64), batch_size = 32, class_mode = "categorical")
```

```python
print(x_train.class_indices)
```

```python
crop_protection.fit_generator(x_train, samples_per_epoch = 800, epochs = 250, validation_data = x_test, nb_val_samples = 200)
```

```python
crop_protection.save("cropprotection.h5")
```
